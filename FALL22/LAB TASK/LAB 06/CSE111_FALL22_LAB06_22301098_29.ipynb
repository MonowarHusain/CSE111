```python
#task2
class Assassin:
    total = 0
    def __init__(self,name,s_rate):
        Assassin.total+=1
        self.name = name
        self.s_rate = s_rate
    def printDetails(self):
        print(f'Name: {self.name}\nSuccess Rate: {self.s_rate}%\nTotal number of Assassin: {Assassin.total}')
    @classmethod
    def failureRate(cls,name,ns_rate):
        return cls(name,100-ns_rate)
    @classmethod
    def failurePercentage(cls,name,x):
        return cls(name,100-(int(x[0:2])))

john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage("Akabane", "10%")
akabane.printDetails()
```

```python
#task5
class Employee:
    def __init__(self,name,wp):
        self.name = name
        self.workingPeriod = wp
    @classmethod
    def employeeByJoiningYear(cls,name,wp):
        return cls(name,2022-wp)
    @staticmethod
    def experienceCheck(ex,gen):
        if gen == 'male':
            if ex<3:
                return 'He is not experienced'
            else:
                return 'He is experienced'
        elif gen == 'female':
            if ex<3:
                return 'She is not experienced'
            else:
                return 'She is experienced'
employee1 = Employee('Dororo', 3)
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, "male"))
print(Employee.experienceCheck(3, "female"))
```

```python
class Laptop:
    laptopCount = 0
    def __init__(self,name,count):
        self.name = name
        self.count = count
        Laptop.laptopCount += count

    @classmethod
    def advantage(cls):
        print('Laptops are portable')
    @staticmethod
    def resetCount():
        Laptop.laptopCount = 0



lenovo = Laptop("Lenovo", 5);
dell = Laptop("Dell", 7);
print(lenovo.name, lenovo.count)
print(dell.name, dell.count)
print("Total number of Laptops", Laptop.laptopCount)
Laptop.advantage()
Laptop.resetCount()
print("Total number of Laptops", Laptop.laptopCount)
```

```python
class Cat:
    Number_of_cats = 0
    def __init__(self,color,status):
        Cat.Number_of_cats +=1
        self.color = color
        self.status = status
    @classmethod
    def no_parameter(cls):
        return cls('White','siting')
    @classmethod
    def first_parameter(cls,color):
        return cls(color,'siting')
    @classmethod
    def second_parameter(cls,status):
        return cls('Grey',status)
    def printCat(self):
        print(f'{self.color} cat is {self.status}')

    def changeColor(self,color):
        self.color=color


print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
#task8
import math
class Cylinder:
    radius = 5
    height = 18
    def __init__(self, rad, heig):
        print(f'Default radius: {Cylinder.radius} and height:{Cylinder.height}')
        Cylinder.radius = float(rad)
        Cylinder.height = float(heig)
        print(f'Updated: radius:{Cylinder.radius} height:{Cylinder.height}')
        self.radius = Cylinder.radius
        self.height = Cylinder.height
    @staticmethod
    def area(rad, heigh):
        print('Area: ', 2 * math.pi* rad* heigh + 2 * math.pi * rad**2 )
    @staticmethod
    def volume(rad, heigh):
        print('Volume: ', math.pi * rad ** 2 * heigh)

    @classmethod
    def swap(cls, r, h):
        new = cls(h, r)
        return new

    @classmethod
    def changeFormat(cls, one):
        det = list(map(float,one.split('-')))
        new = cls(det[0], det[1])
        return new



c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```
